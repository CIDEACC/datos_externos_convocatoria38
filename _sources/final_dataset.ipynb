````python
df_climate = pd.read_csv("data_processed/data_climate.csv")
df_climate.drop(columns=['Unnamed: 0'], inplace=True)

df_climate['COD_DEPTO'] = df_climate['COD_DEPTO'].astype(str).str.zfill(2)
df_climate['COD_MUNI'] = df_climate['COD_MUNI'].astype(str).str.zfill(3)
df_climate['CLAS_CCDGO'] = df_climate['CLAS_CCDGO'].astype(str).str.zfill(9)
df_climate['SETR_CCNCT'] = df_climate['CLAS_CCDGO'].astype(str).str.zfill(9)
df_climate.drop(columns=['CLAS_CCDGO'], inplace=True)
````

````python
df_manzanas.drop_duplicates('SETR_CCNCT', inplace=True)
key = ['COD_MUNI', 'COD_DEPTO', 'MUNICIPIO', 'DEPARTAMENTO', 'SETR_CCNCT']
centros_poblados_clima = df_manzanas.merge(df_climate, on=key, how='inner')
````
